<h3>No. of season vs Release Year</h3>

```python
x=df['duration'].str.split(' ',expand=True)
x=x[(x[1]=='Seasons') |(x[1]=='Season') ]
x.columns=['count','seasons']

```

```python
x=x.sort_values('count')
```

```python
df['no.of seasons']=x['count']
```

```python
data2=df.loc[df['type']!='Movie']
data2.dropna()
```

```python
data2['No. of seasons'] = pd.to_numeric(data2['no.of seasons'])
data2
```

```python
data2['no.of seasons']=data2['no.of seasons'].astype(int)
data2.info()
```

```python
sns.jointplot(data=data2,x='release_year',y='no.of seasons')
plt.xticks(range(1940,2030,10))
plt.ylabel('No. of seasons')
plt.yticks(range(1,12,1))
plt.savefig('images/season_vs_year_released.png', bbox_inches='tight')
```

<h4> There is a increase in number of season with time</h4>
